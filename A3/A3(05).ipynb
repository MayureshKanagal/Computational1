```python
import numpy as np
import time
from matplotlib import pyplot as plt

T_dir_comp=[]
T_lib=[]
start_number=4
end_number=100
numbers = np.arange(start_number, end_number + 1)

for n in range (start_number,end_number+1):
    random_numbers = np.random.rand(n)
    
    start_time1 = time.time()
    dft_comp=[]
    for q in range (0,n):
        w=[]
        for p in range (0,n):
            x=(random_numbers[p]* np.exp(((-1j)*2*np.pi*q *p)/n))/np.sqrt(n)
            w.append(x)
        wq=sum(w)
        dft_comp.append(wq)
        w.clear()
    end_time1 = time.time()
    elapsed_time1 = end_time1 - start_time1
    T_dir_comp.append(elapsed_time1)
#     print(dft_comp)
    


    start_time2 = time.time()
    dft_lib= np.fft.fft(random_numbers,norm='ortho')
    end_time2 = time.time()
    elapsed_time2 = end_time2 - start_time2
    T_lib.append(elapsed_time2)
#     print(dft_lib)

# print(dft_comp)
# print(dft_lib)

plt.plot(numbers,T_dir_comp,"o",color='blue',label='Time taken for direct computation')
plt.plot(numbers,T_lib,"o",color='orange',label='Time taken by numpy')
plt.ylabel('time in sec')
plt.xlabel('Total numbers')
plt.title('Time taken as function of n')
plt.legend()
plt.show()
```
