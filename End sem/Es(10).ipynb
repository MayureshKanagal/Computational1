```python

import numpy as np
import matplotlib.pyplot as plt


x_min = -50 
x_max = 50  
num_samples =1024  
delta_x = (x_max - x_min) / (num_samples-1)  
x = np.linspace(x_min, x_max, num_samples, endpoint=False)

def f_x(x):
    return np.exp(-2 * np.abs(x))

f_x_values = f_x(x)

fft_result = np.fft.fftshift(np.fft.fft(f_x_values, norm='ortho'))
freqs = np.fft.fftshift(np.fft.fftfreq(len(f_x_values), d=1/sampling_freq))

k=(2*np.pi)*freqs
ft_fun =delta_x * np.sqrt(num_samples / (2 * np.pi)) * np.exp(-1j * k * x_min) *fft_result

plt.figure(figsize=(12, 6))

plt.plot(x, f_x_values, label='f(x) = exp(-2*|x|)')
plt.title('Original Function f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# a) Plotting real and Imaginary part
plt.plot(k, np.real(ft_fun), label='|F(k)|')
plt.title('Magnitude of Fourier Transform |F(k)|')
plt.xlabel('Frequency (k)')
plt.ylabel('Magnitude')
plt.legend()
plt.plot(k, np.imag(ft_fun), label='|F(k)|')
plt.title('Magnitude of Fourier Transform |F(k)|')
plt.xlabel('Frequency (k)')
plt.ylabel('Magnitude')
plt.legend()
plt.show()

# b) Overplot the analytic function

def analytic_solution(k):
    return np.sqrt(8/np.pi) / (k**2 + 4)
sol= analytic_solution(k)

plt.plot(k,sol,label='Analytic solution')
plt.plot(k, np.abs(ft_fun), label='Numerical Fourier transform|F(k)|')
plt.title('Comparison of analytic and numerical Fourier Transform |F(k)|')
plt.xlabel('Frequency (k)')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
```

Real Part of Fourier Transform
For an even function, the real part of its Fourier transform is also an even function
   - The real part of the Fourier transform captures the symmetric (even) components of the function.
   
Imaginary Part of the Fourier Transform:
For an even function, the imaginary part of its Fourier transform is an odd function.
   - The imaginary part of the Fourier transform captures the anti-symmetric (odd) components of the function.
   
In summary, for an even function:
- The real part of its Fourier transform captures the even components of the function.
- The imaginary part of its Fourier transform captures the odd components of the function.
