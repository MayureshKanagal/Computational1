```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y, z):
    return z, (2 * t * z - 2 * y + t**3 * np.log(t)) / (t**2)

def exact_sol(t):
    return ((7*t/4) + (t**3/2) * np.log(t) - (3/4) * t**3)

h=0.001
start=1
stop=2
n=(stop-start)/h

x0=1          #initial x
y0=1          # initial y
z0=0          #initial z(y')
t_val=[x0]
y_val=[y0]
z_val=[z0]

for _ in np.arange(n):
    t=t_val[-1]
    z=z_val[-1]
    y=y_val[-1]
    
    y1_val= y + h* f(t,y,z)[0]
    z1_val= z + h* f(t,y,z)[1]
    
    y_val.append(y1_val)
    z_val.append(z1_val)
    t_val.append(t+h)

# print('y_val=',y_val)

act_sol=[]
for i in range(len(t_val)):
    exsol= exact_sol(t_val[i])
    act_sol.append(exsol)
    
# print('Exact Solution=',act_sol)

plt.plot(t_val,y_val,label='Euler Method')
plt.plot(t_val,act_sol,label='Exact solution',linestyle='--')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Comparison of Euler Method with Exact Solution')
plt.legend()
plt.show()
```
