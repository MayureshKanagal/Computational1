using the fact the solve_ivp uses Adaptive RK4

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def fun(t, y):
    return (y**2 + y) / t

t_span = (1, 3)
y0 = -2

abs_tol = 1e-7

sol = solve_ivp(fun, t_span, [y0], method='RK45', atol=abs_tol, rtol=0)

if sol.success:
    print("Solution was successful!")
else:
    print("Solution failed!")


plt.figure(figsize=(10, 6))

plt.plot(sol.t, sol.y[0], label='Solution Curve', color='blue')

plt.scatter(sol.t, sol.y[0], label='Mesh Points', color='red', marker='o')

plt.title("Solution of the ODE with RK45 and Adaptive Step Size Control")
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
