solution (8a)

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

def f1(t,y):
    return(t*np.exp(3*t)-2*y)
def y_exact(t):
    return (1/25) * np.exp(-2 * t) * (1 + np.exp(5 * t) * (-1 + 5 * t))
start = 0
stop= 1
#Total number of interval
n=50

yi=0

sol= scipy.integrate.solve_ivp(f1,t_span=[start,stop],y0=[yi],method='RK45',t_eval=np.linspace(start, stop, n))
print(sol)

analytic_value = y_exact(sol.t)
print(analytic_value)

plt.plot(sol.t,sol.y[0],label='Numerical Solution')
plt.plot(sol.t,analytic_value,label='Analytic Solution', linestyle='--')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Comparison of Numerical and Analytic Solutions')
plt.legend()
plt.grid(True)
plt.show()
```
