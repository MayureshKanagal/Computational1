Solution (9a)

```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt

def fun(x, y):
    return np.vstack((y[1], -np.exp(-2*y[0])))


def bc(ya, yb):                # Function evaluating residuals of the boundary conditions.
    return np.array([ya[0] - 0, yb[0] - np.log(2)])

def exact(x):
    return(np.log(x))
start=1
stop= 2
n=100

x = np.linspace(start, stop,n)

y_guess = np.zeros((2, x.size))
y_guess[0, :] = np.exp(x)  
y_guess[1, :] = np.exp(x)  

sol = solve_bvp(fun, bc, x, y_guess)

print(sol)

x_plot = np.linspace(start, stop,10)
y_sol = sol.sol(x_plot)[0]
plt.plot(x_plot, y_sol, label='y(x)')
plt.plot(x_plot, exact(x_plot), label='Exact Solution', color='green',linestyle='--')
plt.title("Solution of the BVP")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
