```python
import numpy as np
import matplotlib.pyplot as plt
import time

def box_muller(size):
    u1 = np.random.rand(size // 2)
    u2 = np.random.rand(size // 2)
    r = np.sqrt(-2 * np.log(u1))
    theta = 2 * np.pi * u2
    z1 = r * np.cos(theta)
    z2 = r * np.sin(theta)
    return np.concatenate((z1, z2))

num_samples = 10000
start_time = time.time()
random_numbers_box_muller = box_muller(num_samples)
end_time = time.time()
box_muller_time = end_time - start_time
print(f"Time taken by Box-Muller transform to generate 10,000 random numbers: {box_muller_time:.6f} seconds")

plt.hist(random_numbers_box_muller, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(-4, 4, 1000)
gaussian_pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
plt.plot(x, gaussian_pdf, 'r--', linewidth=2, label='Gaussian PDF')

# Add titles and labels
plt.title('Density Histogram of Box-Muller Generated Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
