```python
import numpy as np
import matplotlib.pyplot as plt


a = 1664525
c = 1013904223
m = 2**32
seed = 42

def lcg(seed, a, c, m, size):
    numbers = []
    X = seed
    for _ in range(size):
        X = (a * X + c) % m
        numbers.append(X / m)         # Normalisation
    return numbers

num_samples = 10000
random_numbers = lcg(seed, a, c, m, num_samples)

plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g')

x = np.linspace(0, 1, 100)
plt.plot(x, [1] * len(x), 'r--', linewidth=2, label='Uniform PDF')

plt.title('Density Histogram of LCG Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
