```python
import numpy as np
import matplotlib.pyplot as plt

num_samples = 10000
random_numbers = np.random.rand(num_samples)

plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')

x = np.linspace(0, 1, 100)

plt.plot(x, [1] * len(x), 'r--', linewidth=2, label='Uniform PDF')

plt.title('Density Histogram of np.random.rand() Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
