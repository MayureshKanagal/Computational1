```python
import time
import numpy as np
import matplotlib.pyplot as plt

a = 1664525
c = 1013904223
m = 2**32
seed = 42

def lcg(seed, a, c, m, size):
    numbers = []
    X = seed
    for _ in range(size):
        X = (a * X + c) % m
        numbers.append(X / m)
    return numbers

num_samples = 10000
start_time = time.time()
random_numbers_lcg = lcg(seed, a, c, m, num_samples)
end_time = time.time()
lcg_time = end_time - start_time
print(f"Time taken by LCG to generate 10,000 random numbers: {lcg_time:.6f} seconds")

# Plot the density histogram for LCG
plt.hist(random_numbers_lcg, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')
x = np.linspace(0, 1, 100)
plt.plot(x, [1] * len(x), 'r--', linewidth=2, label='Uniform PDF')
plt.title('Density Histogram of LCG Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Measure the time taken to generate 10,000 random numbers using np.random.rand()
start_time = time.time()
random_numbers_rand = np.random.rand(num_samples)
end_time = time.time()
rand_time = end_time - start_time
print(f"Time taken by np.random.rand() to generate 10,000 random numbers: {rand_time:.6f} seconds")

# Plot the density histogram for np.random.rand()
plt.hist(random_numbers_rand, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')
x = np.linspace(0, 1, 100)
plt.plot(x, [1] * len(x), 'r--', linewidth=2, label='Uniform PDF')
plt.title('Density Histogram of np.random.rand() Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
