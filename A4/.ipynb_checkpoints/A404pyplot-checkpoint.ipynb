```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import os
# Load the data from your text file (replace 'output.txt' with your actual file name)
file_name = f'A404output.txt'  # Adjust the file name as per your requirement
file_path = os.path.join(os.path.expanduser('~'), 'Desktop','Mayu Software','Computational','Computational1','A4', file_name)  # File path on desktop
loaded_data=np.loadtxt(file_path)
# Create a density histogram
plt.hist(loaded_data, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Density Histogram')

# Plot the exponential PDF
x = np.linspace(min(loaded_data), max(loaded_data), 1000)
mu = 0.5
pdf = expon.pdf(x, scale=mu)  # Exponential PDF with mean mu
plt.plot(x, pdf, color='red', label='Exponential PDF')

# Customize labels and title
plt.xlabel('Random Numbers')
plt.ylabel('Density')
plt.title('Density Histogram vs. Exponential PDF')
plt.legend()

# Show the plot
plt.show()
```
